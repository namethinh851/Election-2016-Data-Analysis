```python
url = 'http://elections.huffingtonpost.com/pollster/2016-general-election-trump-vs-clinton.csv'

source = requests.get(url).text

poll_data = StringIO(source)
```

```python
poll_df = pd.read_csv(poll_data)
```

```python
avg = pd.DataFrame(poll_df.mean())

avg.drop('Question Iteration', axis = 0, inplace=True)
avg.drop('Number of Observations', axis = 0, inplace = True)
```

```python
avg
```

```python
std = pd.DataFrame (poll_df.std())
std.drop('Number of Observations', axis = 0, inplace = True)
std.drop('Question Iteration', axis = 0, inplace = True)
```

```python
std.head()
```

```python
avg.plot(yerr = std, kind = 'bar', legend = False)
```

Since Trump and Clinton is very closed to each other, the 'Undecided' can make a big difference in who will win the race

```python
poll_avg = pd.concat([avg, std], axis = 1)
```

```python
poll_avg.columns = ['Average', 'STD']
poll_avg
```

```python
poll_df['Difference'] = (poll_df.Clinton - poll_df.Trump)/100
```

Positive number in the Difference column means Clinton is leading
Negative number in the Difference column means Trump is leading

```python
poll_df = poll_df.groupby(['Start Date'], as_index = False).mean()

poll_df.head()
```

```python
poll_df.plot('Start Date', 'Difference', figsize = (12,4), marker = 'o', linestyle = '-', color = 'purple')
```

Up means people favor Clintin more, Down means people favor Trump more

```python
row_in = 0
xlimit = []

for date in poll_df['Start Date']:
    if date[0:7] == '2016-09':
        xlimit.append(row_in)
        row_in += 1
    else:
        row_in += 1
        
print min(xlimit)
print max(xlimit)
```

```python
Trump_donor = pd.read_csv('TrumpDonation.csv', index_col = False, dtype={'Max. CPC': pd.np.float64})
```

```python
Trump_donor
```

```python
Clinton_donor = pd.read_csv('Clinton.csv', index_col = False, dtype={'Max. CPC': pd.np.float64})
```

```python
Clinton_donor.head()
```

```python
donor = [Trump_donor, Clinton_donor]
```

```python
donor_df = pd.concat(donor)
```

```python
donor_df.head()
```

```python
donor_df['contb_receipt_amt'].value_counts()
```

```python
don_mean = donor_df['contb_receipt_amt'].mean()

don_std = donor_df['contb_receipt_amt'].std()

print 'The average donation was %.2f with a std %.2f' %(don_mean, don_std)
```

```python
top_donor = donor_df['contb_receipt_amt'].copy()

top_donor.sort()

top_donor
```

```python
top_donor = top_donor [top_donor > 0]

top_donor.sort()
top_donor.value_counts().head(10)
```

The top most donation amount

```python
com_don = top_donor[top_donor < 250]

com_don.hist(bins = 100)
```
